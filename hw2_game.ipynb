```python
import pandas as pd
import numpy as np

purchase_data = pd.read_csv("C:/Users/USER/Downloads/purchase_data.csv")
```

## Player Count

```python
player=purchase_data['SN'].unique()
ans1=pd.DataFrame({'Total Player':[len(player)]})
ans1
```

## Purchasing Analysis (Total)

```python
item=purchase_data['Item ID'].unique()
k = str(round(purchase_data['Price'].mean(),2))
price=[]
for s in item:
    mm=purchase_data['Item ID']==s
    price+=[purchase_data[mm]['Price'].unique()]
ans2=pd.DataFrame({'Number of Unique Items':[len(item)],'Average Price':['$'+k],'Number of Purchases':[len(purchase_data)],'Total Revenue':['$'+str(purchase_data['Price'].sum())]})
ans2
```

## Gender Demographics

```python
index=['Male','Female','Other / Non-Disclosed']
ans3=pd.DataFrame(columns=['Percentage of Players','Total Count'])
boy=purchase_data['Gender']=='Male'
male=purchase_data[boy]['SN'].unique()
girl=purchase_data["Gender"]=='Female'
female=purchase_data[girl]['SN'].unique()
other=purchase_data['Gender']=='Other / Non-Disclosed'
others=purchase_data[other]['SN'].unique()
ll=[male,female,others]
for i in range(len(index)):
    ii=index[i]
    ans3.loc[ii]={'Percentage of Players':round((len(ll[i])/len(player))*100,2),'Total Count':len(ll[i])}
ans3
```

## Purchasing Analysis (Gender)

```python
ans4=pd.DataFrame(columns=['Gender','Purchase Count','Average Purchase Price','Total Purchase Value','Avg Purchase Total per Person'])
index,name=['Female','Male','Other / Non-Disclosed'],[girl,boy,other]
ll=[female,male,others]
for i in range(3):
    ans4.loc[i]={'Gender':index[i],
                 'Purchase Count':len(purchase_data[name[i]]),
                 'Average Purchase Price':round(purchase_data[name[i]]['Price'].mean(),5),
                 'Total Purchase Value':round(purchase_data[name[i]]['Price'].sum(),2),
                 'Avg Purchase Total per Person':'$'+str(round(purchase_data[name[i]]['Price'].sum()/len(ll[i]),2))
                }
ans4.set_index('Gender',inplace=False)
```

## Age Demographics

```python
age=[]
for t in player:
    index=purchase_data.index[purchase_data['SN']==t]
    age+=[purchase_data['Age'][index[0]]]
ans5=pd.DataFrame({'Name':player,'age':age})
bins_age=[0,9,14,19,24,29,34,39,100]
mylabel=['<10','10-14','15-19','20-24','24-29','30-34','35-39','40+']
ans5['Age']=pd.cut(ans5['age'],bins_age,labels=mylabel)
ans5
ans5=ans5.rename(columns={'Name':'Percentage of Players','age':'Total Count'})
ans5=ans5.groupby(['Age']).count()
ans5['Percentage of Players']=round((ans5['Percentage of Players']/len(player))*100,2)
ans5
```

## Purchasing Analysis (Age)

```python
ans6=pd.DataFrame({'Purchase Count':purchase_data['Age']})
ans6['Age']=pd.cut(ans6['Purchase Count'],bins_age,labels=mylabel)
ans6=ans6.groupby(['Age']).count()
total_price,total_math,l,ll=[],[],[],[]
for j in range(len(ans6)):
    mas=purchase_data['Age']>bins_age[j]
    mas1=purchase_data[mas]['Age']<=bins_age[j+1]
    total_price+=['$'+str(round(purchase_data[mas][mas1]['Price'].sum(),2))]
    total_math+=[round(purchase_data[mas][mas1]['Price'].sum(),2)]
ans6['Total Purchase Value']=total_price
ans6['price']=total_math
ans6['diff people']=ans5['Total Count']
k=round(ans6['price']/ans6['Purchase Count'],2)
for jj in range(len(k)):
    l.append('$'+str(k[jj]))
ans6['Average Purchase Price']=l
kk=round(ans6['price']/ans6['diff people'],2)
for jj in range(len(kk)):
    ll.append('$'+str(kk[jj]))
ans6['Total per Person']=ll
ans6[['Purchase Count','Average Purchase Price','Total Purchase Value','Total per Person']]
```

## Top Spenders

```python
ans7=pd.DataFrame(columns=['SN','Purchase Count','Average Purchase Price','Total Purchase Value','total_math'])
ave_price,total_price=[],[]
for d in range(len(player)):
    d_=player[d]
    dd=purchase_data['SN']==d_
    ans7.loc[d]={'SN':d_,
                 'Purchase Count':len(purchase_data[dd]),
                 'Average Purchase Price':'$'+str(round(purchase_data[dd]['Price'].sum()/len(purchase_data[dd]),2)),
                 'Total Purchase Value':'$'+str(round(purchase_data[dd]['Price'].sum(),2)),
                 'total_math':round(purchase_data[dd]['Price'].sum(),2)
                }
ans7=ans7.sort_values(by='total_math',ascending=False)
ans7=ans7.set_index('SN')
ans7[['Purchase Count','Average Purchase Price','Total Purchase Value']].head()
```

## Most Popular Items

```python
ans8=pd.DataFrame(columns=['Item ID','Item Name','Purchase Count','Item Price','Total Purchase Value','total_math'])
for f in range(len(item)):
    ff=item[f]
    item_=purchase_data['Item ID']==ff
    ans8.loc[f]={'Item ID':ff,
                 'Item Name':(purchase_data[item_]['Item Name'].unique())[0],
                 'Purchase Count':len(purchase_data[item_]),
                 'Item Price':'$'+str(price[f][0]),
                 'Total Purchase Value':'$'+str(round(purchase_data[item_]['Price'].sum(),2)),
                 'total_math':round(purchase_data[item_]['Price'].sum(),2)
                }
ans8=ans8.sort_values(by='Purchase Count',ascending=False)
ans8=ans8.set_index(['Item ID','Item Name'])
ans8[['Purchase Count','Item Price','Total Purchase Value']].head()
```
